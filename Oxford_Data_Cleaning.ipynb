```python
# Import Dependencies 
import pandas as pd
import numpy as np
```

```python
#Read CSV file
oxford_data = pd.read_csv("Resources/oxford_latest.csv", low_memory=False)
oxford_data
```

```python
oxford_data.update(oxford_data[['RegionName', 'RegionCode']].fillna(0))
oxford = oxford_data[oxford_data.RegionName == 0]
oxford
```

```python
#Save new DataFrame to a csv file
oxford.to_csv("output_data/oxford_all_international.csv", index=False)
```

```python
oxford_df = oxford[(oxford['CountryName'].isin(["Brazil", "China", "Italy", "Japan",\
                                                "New Zealand", "Sweden", "United Kingdom",\
                                                "United States"]))]
oxford_df
```

```python
# Create new dataframe with the 8 countries of interest
oxford_df = pd.DataFrame(oxford_df, columns = [ "CountryName", "Date", "C1_School_closing", "C2_Workplace_closing", \
                                               "C6_Stay_home_requirements", "E1_Income_support", "E3_Fiscal_measures", \
                                               "H6_Facial_coverings", "ConfirmedCases", "ConfirmedDeaths"])
oxford_df = oxford_df.dropna()
oxford_df
```

```python
#Save new DataFrame to a csv file
oxford_df.to_csv("output_data/oxford_clean.csv", index=False)
```
